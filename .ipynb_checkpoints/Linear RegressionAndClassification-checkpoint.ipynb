```python
import plotly.express as px
import pandas as pd
import numpy as np

data = pd.read_csv('C:\\Users\\HP\\Documents\\TPexam\\Cleaned_ds_salaries.csv')
```

**TEST/TRAIN split**

```python
#lets set our target or 'y' as the Savings and then remove it from the dataset (X).
#y = f(X)
y = data['Cumulative_Savings10%_of_their_salary_annually']
X = data.drop(['Cumulative_Savings10%_of_their_salary_annually'],axis=1)
# Convert categorical variables to dummy variables
X = pd.get_dummies(X, drop_first=True)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1)
```

## Linear Regression

```python
#Set model (Am going to use LinearRegression.)
from sklearn.linear_model import LinearRegression
```

```python
#Fitting the model to the training data.
model = LinearRegression()
```

```python
model.fit(X_train,y_train)
```

```python
#using the fitted model to predict the test Savings and calculate the accuracy by useing "score".
model.score(X_test,y_test)
```

```python
#Grabbing a sample (row) from the test set so that i can do a prediction.
X_demo = X_test.iloc[5].values.reshape(1,-1)
y_demo = y_test.iloc[5]
```

```python
y_pred = model.predict(X_demo)
print('Prediction: ' + str(y_pred[0]))
print('Measured: ' + str(y_demo))
```

## Classification

```python
X = data.iloc[:,0:4]
y = data['Cumulative_Savings10%_of_their_salary_annually']
```

```python
#Normalise X
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
scaler.fit(X)
X_norm = scaler.transform(X)
```

```python
from sklearn import preprocessing
le = preprocessing.LabelEncoder()
le.fit(y)
y_transf = le.transform(y)
print(y_transf)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_norm, y_transf, test_size=0.3, random_state=15)
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
from sklearn.model_selection import cross_validate
```

```python
#Tune the hyperparameters on the training dataset using cross_validation.
model = RandomForestClassifier(n_estimators=1)

model_trained = cross_validate(model, X_train, y_train, cv=3)
result = np.mean(model_trained['test_score'])
print(result)
```

```python
# Fitting the model
model_trained = model.fit(X_train, y_train)
pred_cl = model_trained.predict(X_test)
```

```python
#The F1-Score
from sklearn.metrics import f1_score
f1 = f1_score(y_test, pred_cl, average='weighted')
print(f1)
```
