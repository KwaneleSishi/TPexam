```python
#Renaming the value in the columns for better understanding
data['experience_level'] = data['experience_level'].replace({
    'SE': 'Senior',
    'EN': 'Entry level',
    'EX': 'Executive level',
    'MI': 'Mid/Intermediate level',
})

data['employment_type'] = data['employment_type'].replace({
    'FL': 'Freelancer',
    'CT': 'Contractor',
    'FT' : 'Full-time',
    'PT' : 'Part-time'
})
data['company_size'] = data['company_size'].replace({
    'S': 'SMALL',
    'M': 'MEDIUM',
    'L' : 'LARGE',
})
data['remote_ratio'] = data['remote_ratio'].astype(str)
data['remote_ratio'] = data['remote_ratio'].replace({
    '0': 'On-Site',
    '50': 'Half-Remote',
    '100' : 'Full-Remote',
})
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# List of numerical features to visualize
numerical_features = ['salary', 'Cumulative_Savings10%_of_their_salary_annually']

# Creating boxplots for numerical features
plt.figure(figsize=(15, 5))  # Adjust figure size for two plots

for i, feature in enumerate(numerical_features, 1):
    plt.subplot(1, 2, i)  # 1 row, 2 columns, iterate through each feature
    sns.boxplot(y=data[feature])
    plt.title(f'Boxplot of {feature}', fontsize=14)
    plt.xlabel('')  # No need for x-axis labels on a boxplot
    plt.ylabel(feature)  # Label y-axis as the feature name

plt.tight_layout()  # Adjust layout for no overlap
plt.show()
```

```python
job_titles = ['Data Analyst', 'Data Scientist', 'ML Engineer', 'Applied Scientist']
filtered_data = data[data['job_title'].isin(job_titles)]
```

```python
# Count occurrences of each job title for the pie chart
job_counts = filtered_data['job_title'].value_counts()

# Create a pie chart
plt.figure(figsize=(10, 6))
plt.subplot(1, 2, 1)  # Create a subplot for the pie chart
job_counts.plot(kind='pie', autopct='%1.1f%%', startangle=90)
plt.title('Pie Chart of Job Titles')
plt.ylabel('')  

total_salary = filtered_data.groupby('job_title')['salary_in_usd'].sum()

# Create a bar graph
plt.subplot(1, 2, 2)  # Create a subplot for the bar graph
total_salary.plot(kind='bar', color='skyblue')
plt.title('Total Salary by Job Title')
plt.xlabel('Job Title')
plt.ylabel('Total Salary in USD')
plt.xticks(rotation=45)  

plt.tight_layout()  
plt.show()
```
